```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
m = pow(2, 32)
a = 1664525
c = 1013904223
state = 1  # seed

def lcg_random():
    global state
    state = (a * state + c) % m
    return state / m
```

```python
points = [lcg_random() for i in range(10000)]
```

```python
mu = 1
sigma = 1

X = []

for i in range(10000):
  p1 = random.choice(points)
  p2 = random.choice(points)
  X.append(mu + np.sqrt(-2*np.log(p1))*np.cos(2*np.pi*p2) * sigma)
```

```python
n, bins, patches = plt.hist(X, bins=50, density=True, alpha=0.7) # Projects X-values to histogram
```

```python
from scipy.stats import norm

mu = 1.0
sigma = 1.0

def target_pdf(x):
    return np.exp(-0.5 * ((x - mu) / sigma)**2)

h = 2.0
n_iter = 50000
burn_in = 5000

x_current = 0.0
samples = []
accepted = 0

for t in range(n_iter):
    delta = np.random.uniform(-h, h)
    x_proposal = x_current + delta

    alpha = min(1.0, target_pdf(x_proposal) / target_pdf(x_current))

    u = np.random.uniform(0, 1)

    if u <= alpha:
        x_current = x_proposal
        if t >= burn_in:
            accepted += 1

    if t >= burn_in:
        samples.append(x_current)

samples = np.array(samples)

print(f"Среднее выборки: {samples.mean():.3f} (ожидается 1.0)")
print(f"Стандартное отклонение: {samples.std():.3f} (ожидается 1.0)")
print(f"Доля принятых предложений: {accepted / len(samples):.2f}")

n, bins, patches = plt.hist(samples, bins=50, density=True, alpha=0.7) # Projects samples to histogram

x_plot = np.linspace(-3, 5, 100)
plt.plot(x_plot, norm.pdf(x_plot, mu, sigma), 'r-', lw=1, label='N(1,1)')
plt.show()
```

```python
d = 3
bins_count = 50
n = 10000
```

```python
def lcg_random():
    global state
    state = (a * state + c) % m
    return state / m

bins = [0 for i in range(bins_count + 1)]

for i in range(n):
  vector = [lcg_random() for i in range(d)]
  M = max(vector)
  k = math.floor(bins_count * pow(M, d)) + 1
  bins[k] += 1

expected = n / bins_count
ksi_squared = sum((count - expected)**2 / expected for count in bins[1:]) # skip unused bins[0]
p_value = 1 - chi2.cdf(ksi_squared, df=bins_count - 1)

print(bins)
print(ksi_squared)
print(p_value)

if p_value > 0.05:
    print("Accept: uniform")
else:
    print("Reject: non-uniform")
```
