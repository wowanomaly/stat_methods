```python
import random
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations

def linear_func(a, b, x):
  return a * x + b + np.random.normal(0, 1, size=x.shape)

def linear_func_clear(a, b, x):
  return a * x + b

random_xs = [np.float64(random.randint(1, 100)) for i in range(100)]

points = [(np.float64(i), np.float64(linear_func(1, 1, i))) for i in random_xs]

add_points = []

for i in range(10):
  add_points.append((np.float64(random.randint(1, 100)), np.float64(random.randint(1, 100))))

plt.scatter([i[0] for i in points], [i[1] for i in points])
plt.scatter([i[0] for i in add_points], [i[1] for i in add_points])
plt.show()

all_points = points + add_points
combs = list(combinations(all_points, 2))

params = []
for comb in combs:
  try:
    a = (comb[0][1] - comb[1][1])/(comb[0][0] - comb[1][0])
    b = comb[0][1] - a * comb[0][0]
    if a != np.nan and a != np.inf and b != np.nan and b != np.inf:
      params.append((a, b))
  except Exception as e:
    pass
```

```python
plt.scatter([i[0] for i in params], [i[1] for i in params], s=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```

```python
a_values = np.array([i[0] for i in params])
n, bins, patches = plt.hist(a_values, bins=200, alpha=0.7) # Projects a-values to histogram
counts = [p.get_height() for p in patches]

high_count_bin_indices = []
for i, patch in enumerate(patches):
    if counts[i] > 1000:
        high_count_bin_indices.append(i)

high_count_bin_indices = np.array(high_count_bin_indices)

bin_indices = np.digitize(a_values, bins) - 1
mask = np.isin(bin_indices, high_count_bin_indices)
high_count_points = a_values[mask]

a_cool = np.mean(high_count_points)
a_cool
```

```python
b_values = [i[1] for i in params]
n, bins, patches = plt.hist(b_values, bins=200, alpha=0.7) # Projects b-values to histogram
counts = [p.get_height() for p in patches]

high_count_bin_indices = []
for i, patch in enumerate(patches):
    if counts[i] > 1000:
        high_count_bin_indices.append(i)

high_count_bin_indices = np.array(high_count_bin_indices)

bin_indices = np.digitize(a_values, bins) - 1
mask = np.isin(bin_indices, high_count_bin_indices)
high_count_points = a_values[mask]

b_cool = np.mean(high_count_points)
b_cool
```

```python
plt.scatter([i[0] for i in points], [i[1] for i in points])
plt.scatter([i[0] for i in add_points], [i[1] for i in add_points])
plt.plot([i for i in range(1, 100)], [linear_func_clear(a_cool, b_cool, np.float64(i)) for i in range(1, 100)])
plt.show()
```
