```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

fs = 1000
T = 1.0
t = np.linspace(0, T, int(fs*T))

s_clean = np.sin(8 * np.pi * t)
window = np.heaviside(t - 0.5, 1) - np.heaviside(t - 0.75, 1)
s_clean += 0.5 * np.sin(40 * np.pi * t) * window

np.random.seed(42)
noise = 0.25 * np.random.normal(0, 1, size=t.shape)
s_noisy = s_clean + noise

plt.figure(figsize=(12, 6))
plt.title('Зашумлённый и исходный сигналы')
plt.plot(t, s_noisy, label='Зашумлённый сигнал')
plt.plot(t, s_clean, label='Исходный сигнал')
plt.legend()
plt.grid()
plt.show()
```

```python
wavelet = 'db4'
level = 7
coeffs = pywt.wavedec(s_noisy, wavelet, level=level)
sigma = np.median(np.abs(coeffs[-1])) / 0.6745
threshold = sigma * np.sqrt(2 * np.log(len(s_noisy)))
coeffs_th = [coeffs[0]] + [pywt.threshold(c, threshold, mode='soft') for c in coeffs[1:]]
s_denoised = pywt.waverec(coeffs_th, wavelet)

plt.figure(figsize=(12, 6))
plt.title('Шумоподавление')
plt.plot(t, s_noisy, label='Зашумлённый сигнал')
plt.plot(t, s_denoised, label='После шумоподавления')
plt.legend()
plt.grid()
plt.show()
```

```python
# Реконструкция низкочастотной части
coeffs_low = [coeffs[0]] + [np.zeros_like(c) for c in coeffs[1:]]
s_low = pywt.waverec(coeffs_low, wavelet)

# Реконструкция среднечастотной части
coeffs_mid = [np.zeros_like(coeffs[0])]
for i in range(1, level+1):
    if i in [2, 3]:
        coeffs_mid.append(coeffs[i])
    else:
        coeffs_mid.append(np.zeros_like(coeffs[i]))
s_mid = pywt.waverec(coeffs_mid, wavelet)

# Обрезаем до исходной длины
s_low = s_low[:len(t)]
s_mid = s_mid[:len(t)]

# Визуализация (сравнение низкочастотной части)
plt.figure(figsize=(12, 8))
plt.title('Низкочастотная компонента')
plt.plot(t, np.sin(8*np.pi*t), label='Истинная локальная sin(8*pi*t)')
plt.plot(t, s_low, label='Восстановленная низкочастотная часть')
plt.legend()
plt.grid()
plt.show()

# Визуализация (сравнение среднечастотной части)
plt.figure(figsize=(12, 8))
plt.title('Среднечастотная компонента')
plt.plot(t, 0.5*np.sin(40*np.pi*t), label='Истинная локальная 0.5*sin(40*pi*t)')
plt.plot(t, s_mid, label='Восстановленная среднечастотная часть')
plt.xlim(0.5, 0.75)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.title('Сравнение исходного и очищенного сигналов')
plt.plot(t, s_clean, label='Исходный без шума')
plt.plot(t, s_denoised, label='Очищенный')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

np.random.seed(42)
N = 1024
m = np.linspace(1.05, 1.25, N)
dm = m[1] - m[0]
background_true = (1200+ 1800*(m - 1.05)+ 250*np.exp(-((m - 1.18)/0.05)**2))
m0 = 1.115
sigma = 0.0025
ampl = 180
signal_true = ampl * np.exp(-0.5*((m - m0)/sigma)**2)
spectrum_true = background_true + signal_true
spectrum_noisy = np.random.poisson(np.clip(spectrum_true, 0, None))
plt.figure(figsize=(12, 5))
plt.plot(m, spectrum_noisy, label='Noisy spectrum', linewidth=1)
plt.plot(m, background_true, '--', label='True background', linewidth=2)
plt.plot(m, signal_true, '--', label='True signal', linewidth=2)
plt.xlabel(r'Invariant mass $m$')
plt.ylabel('Counts')
plt.title('Synthetic mass spectrum: background + narrow signal + noise')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
wavelet = 'db4'
level = 5
coeffs = pywt.wavedec(spectrum_noisy, wavelet, level=level)
sigma = np.median(np.abs(coeffs[-1])) / 0.6745
threshold = sigma * np.sqrt(2 * np.log(len(s_noisy)))
coeffs_th = [coeffs[0]] + [pywt.threshold(c, threshold, mode='soft') for c in coeffs[1:]]
s_denoised = pywt.waverec(coeffs_th, wavelet)
true_sig = s_denoised - background_true
plt.figure(figsize=(12, 6))
plt.title('Восстановленный сигнал после вейвлет-преобразования и вычитания фона')
plt.plot(m, true_sig)
plt.ylim(0, 1750)
plt.grid()
plt.show()
```
