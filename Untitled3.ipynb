```python
import numpy as np
from scipy.optimize import least_squares
import matplotlib.pyplot as plt

A=2.0
beta=0.25
omega=4.5
phi=0.7
c=0.2

sigma=0.15
eps=0
t_min=0
t_max=10
N=501

t = np.linspace(t_min, t_max, N)
y_data = A*np.exp(-beta*t)*np.sin(omega*t + phi) + c + np.random.normal(0, sigma, size=t.shape)

def y_no_noise(t, theta):
  A, beta, omega, phi, c = theta
  return A*np.exp(-beta*t)*np.sin(omega*t + phi) + c

def residuals(theta):
    return y_data - y_no_noise(t, theta)

# theta = np.array([A, beta, omega, phi, c])
initial_theta1 = np.array([1.5, 0.19, 3.9, 0.5, 0.1])
initial_theta2 = np.array([1.6, 0.17, 4.9, 0.9, 0.3])
initial_theta3 = np.array([2.5, 0.3, 4.3, 0.6, 0.15])
initial_theta4 = np.array([1.7, 0.15, 4.7, 0.8, 0.25])
initial_theta5 = np.array([2.3, 0.23, 3.7, 0.4, 0.17])

result1 = least_squares(residuals, initial_theta1, method='lm') # using Levenberg-Marquardt algorithm which is well-suited for nonlinear least squares problems like this damped oscillator fitting
theta_star1 = result1.x # extract optimized parameters theta* = [A, beta, omega, phi, c]
print(theta_star1)

result2 = least_squares(residuals, initial_theta2, method='lm') # using Levenberg-Marquardt algorithm which is well-suited for nonlinear least squares problems like this damped oscillator fitting
theta_star2 = result2.x # extract optimized parameters theta* = [A, beta, omega, phi, c]
print(theta_star2)

result3 = least_squares(residuals, initial_theta3, method='lm') # using Levenberg-Marquardt algorithm which is well-suited for nonlinear least squares problems like this damped oscillator fitting
theta_star3 = result3.x # extract optimized parameters theta* = [A, beta, omega, phi, c]
print(theta_star3)

result4 = least_squares(residuals, initial_theta4, method='lm') # using Levenberg-Marquardt algorithm which is well-suited for nonlinear least squares problems like this damped oscillator fitting
theta_star4 = result4.x # extract optimized parameters theta* = [A, beta, omega, phi, c]
print(theta_star4)

result5 = least_squares(residuals, initial_theta5, method='lm') # using Levenberg-Marquardt algorithm which is well-suited for nonlinear least squares problems like this damped oscillator fitting
theta_star5 = result5.x # extract optimized parameters theta* = [A, beta, omega, phi, c]
print(theta_star5)
```
