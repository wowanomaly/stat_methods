```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, accuracy_score, confusion_matrix, ConfusionMatrixDisplay

def points_in_disk(center, radius, num_points):
    cx, cy = center
    points = []
    for _ in range(num_points):
        r = np.sqrt(np.random.random()) * radius
        theta = np.random.random() * 2 * np.pi
        x = cx + r * np.cos(theta)
        y = cy + r * np.sin(theta)
        points.append([x, y])
    return np.array(points)

def points_in_annulus(center, r_inner, r_outer, num_points):
    cx, cy = center
    points = []
    for _ in range(num_points):
        r = np.sqrt(np.random.random() * (r_outer**2 - r_inner**2) + r_inner**2)
        theta = np.random.random() * 2 * np.pi
        x = cx + r * np.cos(theta)
        y = cy + r * np.sin(theta)
        points.append([x, y])
    return np.array(points)

def group_points(points, center_x=5.0):
    groups = []
    x_rel = points[:, 0] - center_x

    # Define intervals (left inclusive, right exclusive except last)
    intervals = [(-2.5, -1.5), (-1.5, 0.0), (0.0, 1.5), (1.5, 2.5)]

    for i, (low, high) in enumerate(intervals, 1):
        if i < 4:
            mask = (x_rel >= low) & (x_rel < high)
        else:
            mask = (x_rel >= low) & (x_rel <= high)   # include right boundary for last group

        group = points[mask]
        groups.append(group)
        print(f"Group {i}: {len(group)} points")

    return groups

def plot_groups(groups, center=(5,5), radii=[1.0,1.5,2.5]):
    colors = ['red', 'green', 'blue', 'orange']
    plt.figure(figsize=(6,6))
    plt.title('Points grouped by relative x')
    for grp, color in zip(groups, colors):
        plt.scatter(grp[:,0], grp[:,1], s=3, color=color)
    for r in radii:
        circle = plt.Circle(center, r, fill=False, edgecolor='black', linestyle='--')
        plt.gca().add_patch(circle)
    plt.scatter(*center, color='black', marker='x')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()

def assign_true_labels(points, center_x=5.0):
    x_rel = points[:, 0] - center_x
    labels = np.zeros(len(points), dtype=int)

    # Intervals (left inclusive, right exclusive except last)
    labels[(x_rel >= -2.5) & (x_rel < -1.5)] = 0
    labels[(x_rel >= -1.5) & (x_rel < 0.0)] = 1
    labels[(x_rel >= 0.0) & (x_rel < 1.5)] = 2
    labels[(x_rel >= 1.5) & (x_rel <= 2.5)] = 3
    return labels


def dumb_classifier(train_labels):
    # Count frequencies of each class
    unique, counts = np.unique(train_labels, return_counts=True)
    class_freq = dict(zip(unique, counts))
    majority_class = unique[np.argmax(counts)]
    print(f"Class frequencies: {class_freq}")
    print(f"Majority class (always predicted): {majority_class}")
    print("\n\n\n")

    # Predict majority class for all points
    predictions = np.full_like(train_labels, majority_class)
    return predictions

def assign_true_binary_labels(points, center_x=5.0, inner_rad=1.0):
    x = points[:, 0]
    blue_mask = (x >= center_x - inner_rad) & (x <= center_x + inner_rad)
    return blue_mask.astype(int)   # 1 for Blue, 0 for Red

def blind_method(y_true):
    p_blue = np.mean(y_true)
    print(f"Proportion of Blue points: {p_blue:.4f}")
    s = np.random.uniform(0, 1, size=len(y_true))
    y_pred = (s < p_blue).astype(int)
    return y_pred


# Parameters
center = (5.0, 5.0)
r1 = 1.0
r2 = 1.5
r3 = 2.5
total_points = 1000
n = total_points // 5   # 200 points inside first circle
m = 4 * n               # 800 points between second and third circle

# Generate points
disk_points = points_in_disk(center, r1, n)
annulus_points = points_in_annulus(center, r2, r3, m)

# Combine and optionally shuffle
all_points = np.vstack([disk_points, annulus_points])
np.random.shuffle(all_points)

# Plot
plt.figure(figsize=(6,6))
plt.title('1000 Points in Concentric Circles')
plt.scatter(all_points[:,0], all_points[:,1], s=3)

# Draw circles for reference
for r in [r1, r2, r3]:
    circle = plt.Circle(center, r, fill=False, edgecolor='red', linestyle='--')
    plt.gca().add_patch(circle)

plt.scatter(*center, color='red', marker='x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

plot_groups(group_points(all_points, center_x=5.0))

# 1. Generate points
points = all_points

# 2. Assign true labels
true_labels = assign_true_labels(points)
print("True label distribution:")
for i in range(4):
    print(f"  Group {i+1} (label {i}): {np.sum(true_labels == i)} points")
print("\n\n\n")

# 3. Dumb classifier: always predict the majority class
predictions = dumb_classifier(true_labels)

# 4. Compute precision and recall
precision = precision_score(true_labels, predictions, average='micro')
recall = recall_score(true_labels, predictions, average='micro')
accuracy = accuracy_score(true_labels, predictions)

print("--- Dumb Classifier Performance ---")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"Accuracy:  {accuracy:.4f}")
print("\n\n\n")

# 5. Confusion matrix
cm = confusion_matrix(true_labels, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1,2,3])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix: Dumb Classifier (always predicts majority class)")
plt.show()

print("\n\n\n")

# True binary labels (Blue = x between 4 and 6)
y_true = assign_true_binary_labels(points, center_x=5.0, inner_rad=1.0)
n_blue = np.sum(y_true)
n_red = len(y_true) - n_blue
print(f"True labels: Blue = {n_blue} points, Red = {n_red} points")

# Blind method predictions
y_pred = blind_method(y_true)
print("\n\n\n")

# Metrics
precision = precision_score(y_true, y_pred, zero_division=0)
recall = recall_score(y_true, y_pred, zero_division=0)
accuracy = accuracy_score(y_true, y_pred)

print("--- Blind Method (Random Classifier) Performance ---")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"Accuracy:  {accuracy:.4f}")

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
disp.plot(cmap='Blues')
plt.title("Blind Method: Random Predictions")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Helper function for covariance matrix (consistent with notebook logic)
def get_cov_matrix(s1, s2, rho):
    return [[s1**2, rho * s1 * s2], [rho * s1 * s2, s2**2]]

# 1. Data Generation (Using Case 4 logic: different covariances)
n_samples = 300
mu_p = np.array([1, 1])
mu_m = np.array([-1, -1])
rho = 0.5

cov_p = get_cov_matrix(1, 1, rho)
cov_m = get_cov_matrix(1, 1, -rho)

X_p = np.random.multivariate_normal(mu_p, cov_p, n_samples)
X_m = np.random.multivariate_normal(mu_m, cov_m, n_samples)

X = np.vstack([X_p, X_m])
y = np.concatenate([np.ones(n_samples), np.zeros(n_samples)])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Apply KNN method
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# 3. Apply Decision Tree using threshold Gini
dt = DecisionTreeClassifier(criterion='gini', random_state=42, max_depth=3)
dt.fit(X_train, y_train)

# 4. Apply Gradient Boosting
gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_train, y_train)

# 5. Visualization of Decision Boundaries
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00'])

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
models = [("KNN (k=5)", knn), ("Decision Tree (Gini)", dt), ("Gradient Boosting", gb)]

for ax, (name, clf) in zip(axes, models):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.5)
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cmap_bold, edgecolors='k', s=50)
    ax.set_title(f"{name}")
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')

plt.tight_layout()
plt.show()

# 6. Evaluate & Print Metrics
models = {
    "KNN": knn,
    "Decision Tree (Gini)": dt,
    "Gradient Boosting": gb
}

print("--- Model Performance ---")
for name, clf in models.items():
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)

    print(f"{name}:")
    print(f"  Accuracy:  {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall:    {rec:.4f}\n")
```
