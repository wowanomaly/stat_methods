```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.tree import DecisionTreeClassifier
```

```python
np.random.seed(42)
```

```python
cluster1 = np.random.randn(250, 2) * 0.75 + np.array([0.0, 0.0])
cluster2 = np.random.randn(120, 2) * 0.25 + np.array([3.0, 2.0])
cluster3 = np.random.randn(40, 2) * 0.12 + np.array([-2.5, 2.5])
cluster4 = np.random.randn(80, 2) * 0.30 + np.array([2.5, -2.5])
noise = np.random.uniform(low=-4.5, high=4.5, size=(40, 2))
X=np.vstack([cluster1, cluster2, cluster3, cluster4, noise])
```

```python
# True labels (for visualization)
true_labels = np.array([0]*250 + [1]*120 + [2]*40 + [3]*80 + [4]*40)
dt_labels = np.array([0]*250 + [0]*120 + [0]*40 + [0]*80 + [0]*40)

# ------------------------------------------------------------
# 2. Plot raw data with true colors
# ------------------------------------------------------------
plt.figure(figsize=(8,6))
scatter = plt.scatter(X[:,0], X[:,1], c=true_labels, cmap='tab10', s=10, alpha=0.7)
plt.title('Raw Data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# ------------------------------------------------------------
# 3. K-Means (k=4)
# ------------------------------------------------------------
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=kmeans_labels, cmap='tab10', s=10, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            marker='X', s=200, c='red', edgecolors='black', label='centroids')
plt.title('K-Means (k=4)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# ------------------------------------------------------------
# 4. DBSCAN (eps and min_samples chosen by trial)
# ------------------------------------------------------------
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

plt.figure(figsize=(8,6))
scatter = plt.scatter(X[:,0], X[:,1], c=dbscan_labels, cmap='tab10', s=10, alpha=0.7)
plt.title('DBSCAN (eps=0.5, min_samples=5)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# ------------------------------------------------------------
# 5. Gaussian Mixture Model (4 components)
# ------------------------------------------------------------
gmm = GaussianMixture(n_components=4, random_state=42)
gmm_labels = gmm.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=gmm_labels, cmap='tab10', s=10, alpha=0.7)
plt.title('GMM (4 components)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# ------------------------------------------------------------
# 6. Decision Tree (supervised, using same labels)
# ------------------------------------------------------------
# Train a decision tree classifier on the whole dataset (just to see boundaries)
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
dt.fit(X, dt_labels)

# Create a mesh to plot decision boundaries
x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
y_min, y_max = X[:,1].min() - 0.5, X[:,1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))
Z = dt.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='tab10')
plt.scatter(X[:,0], X[:,1], c=true_labels, cmap='tab10', s=10, alpha=0.7, edgecolors='k')
plt.title('Decision Tree (trained on same labels) – Decision Boundaries')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
