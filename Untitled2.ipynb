```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_func(a, b, x):
  return a * x + b + np.random.normal(0, 1, size=x.shape)
```

```python
random_xs = [np.float64(random.randint(1, 100)) for i in range(100)]
points = [(np.float64(i), np.float64(linear_func(1, 1, i))) for i in random_xs]
```

```python
add_points = []
```

```python
for i in range(10):
  add_points.append((np.float64(random.randint(1, 100)), np.float64(random.randint(1, 100))))
```

```python
plt.scatter([i[0] for i in points], [i[1] for i in points])
plt.scatter([i[0] for i in add_points], [i[1] for i in add_points])
```

```python
all_points = points + add_points
```

```python
x_vals = np.array([p[0] for p in all_points])
y_vals = np.array([p[1] for p in all_points])
```

```python
n = len(points)

sum_x = np.sum(x_vals)
sum_y = np.sum(y_vals)
sum_xy = np.sum(x_vals * y_vals)
sum_xx = np.sum(x_vals * x_vals)

a_hat_manual = (n * sum_xy - sum_x * sum_y) / (n * sum_xx - sum_x * sum_x)
b_hat_manual = (sum_y - a_hat_manual * sum_x) / n

print("Manual estimates:")
print(f"  a = {a_hat_manual:.4f}  (true a = 1)")
print(f"  b = {b_hat_manual:.4f}  (true b = 1)")
```

```python
x_line = np.linspace(min(x_vals), max(x_vals), 100)
y_line = a_hat_manual * x_line + b_hat_manual

plt.scatter([i[0] for i in points], [i[1] for i in points])
plt.scatter([i[0] for i in add_points], [i[1] for i in add_points])
plt.plot(x_line, y_line, color='r')
```

```python
x = [i[0] for i in points] + [i[0] for i in add_points]
y = [i[1] for i in points] + [i[1] for i in add_points]

X = np.c_[np.ones_like(x), x]
beta = np.linalg.lstsq(X, y, rcond=None)[0]
weights = np.ones(len(x))

for i in range(5):
    W = np.diag(weights)
    beta_new = np.linalg.inv(X.T @ W @ X) @ (X.T @ W @ y)
    residuals = y - X @ beta_new
    weights = 1 / (residuals**2 + 2)
    beta = beta_new
```

```python
x_line = np.linspace(min(x_vals), max(x_vals), 100)
y_line = beta[1] * x_line + beta[0]

plt.scatter([i[0] for i in points], [i[1] for i in points])
plt.scatter([i[0] for i in add_points], [i[1] for i in add_points])
plt.plot(x_line, y_line, color='r')
```
