```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def percentile_manual(data, p):
    n = len(data)
    if n == 0:
        return None
    sorted_data = sorted(data)
    index = (p / 100) * (n - 1)
    i = int(index)
    f = index - i
    if i == n - 1:
        return sorted_data[i]
    return sorted_data[i] + f * (sorted_data[i + 1] - sorted_data[i])

x = [4.92, 5.11, 5.03, 4.88, 5.27, 5.14, 4.95, 5.06, 5.18, 4.97,
     5.09, 4.91, 5.22, 5.04, 5.15, 4.99, 5.08, 5.13, 4.93, 5.17,
     5.01, 4.96, 5.19, 5.12, 4.90, 5.07, 5.16, 4.98, 5.10, 5.05,
     5.21, 4.94, 5.00, 5.23, 5.02, 4.89, 5.20, 5.24, 4.97, 5.06,
     5.18, 4.95, 5.11, 5.03, 4.92, 5.14, 5.08, 5.17, 4.96, 5.12]

n = len(x)
B = 10000
boot_means = []

for _ in range(B):
    sample = np.random.choice(x, size=n, replace=True)
    boot_means.append(np.mean(sample))

ci_lower = np.percentile(boot_means, 2.5)
ci_upper = np.percentile(boot_means, 97.5)

print(f"95% доверительный интервал: [{ci_lower:.3f}, {ci_upper:.3f}]")

ci_lower_manual = percentile_manual(boot_means, 2.5)
ci_upper_manual = percentile_manual(boot_means, 97.5)

print(f"95% доверительный интервал: [{ci_lower_manual:.3f}, {ci_upper_manual:.3f}]")
```

```python
n, bins, patches = plt.hist(boot_means, bins=50, density=True, alpha=0.7) # Projects boot_means to histogram
```

```python
n = len(x)
x_mean = np.mean(x)
x_std = np.std(x, ddof=1)
t_critical = stats.t.ppf(0.975, df=n-1)
t_ci_lower = x_mean - t_critical * (x_std / np.sqrt(n))
t_ci_upper = x_mean + t_critical * (x_std / np.sqrt(n))
print(f"95% t-доверительный интервал: [{t_ci_lower:.3f}, {t_ci_upper:.3f}]")
```
