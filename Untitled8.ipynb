```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# ---------- Функции и их точные интегралы ----------
def f1(x):
    return 4.0 / (1.0 + x**2)

def F1(x):
    """Первообразная f1 (с точностью до константы)"""
    return 4.0 * np.arctan(x)

def f2(x):
    # избегаем вырождения на границах – используем осторожное вычисление
    # x ожидается в (0, pi)
    sinx = np.sin(x)
    # ctg(x) = cos(x)/sin(x)
    return (sinx ** x) * (np.log(sinx) + x * np.cos(x) / sinx)

def F2(x):
    """Первообразная f2"""
    return np.sin(x) ** x

# Точные значения интегралов
a1, b1 = 0.0, 1.0
I1_exact = F1(b1) - F1(a1)   # pi

# Для второй функции формально a=0, b=pi, но в 0 логарифмическая особенность.
# Сдвигаем границы на малую величину, чтобы методы работали устойчиво.
eps = 1e-12
a2, b2 = eps, np.pi - eps
I2_exact = F2(b2) - F2(a2)   # ≈ -1.0

# ---------- Вспомогательные вычислители ----------
def rectangle(f, a, b, M, mode='left'):
    """Интеграл методом прямоугольников с M отрезками.
       mode: 'left', 'right', 'mid'."""
    h = (b - a) / M
    if mode == 'left':
        x = np.linspace(a, b - h, M)
    elif mode == 'right':
        x = np.linspace(a + h, b, M)
    elif mode == 'mid':
        x = np.linspace(a + h/2, b - h/2, M)
    else:
        raise ValueError("mode must be 'left', 'right' or 'mid'")
    return h * np.sum(f(x))

def tree_integral(f, a, b, M, N_train=5000):
    """
    Интеграл с помощью регрессионного дерева.
    f – функция, a,b – границы, M – желаемое число листьев (отрезков).
    N_train – размер равномерной обучающей выборки (N_train >> M).
    """
    # Обучающая выборка
    X_train = np.linspace(a, b, N_train).reshape(-1, 1)
    y_train = f(X_train.ravel())

    # Дерево с ограничением числа листьев
    tree = DecisionTreeRegressor(max_leaf_nodes=M, min_samples_leaf=1)
    tree.fit(X_train, y_train)

    # Извлекаем интервалы и средние значения из листьев
    t = tree.tree_
    leaves = []

    def traverse(node_id, low, high):
        if t.children_left[node_id] == -1:   # лист
            val = t.value[node_id][0][0]
            leaves.append((low, high, val))
        else:
            th = t.threshold[node_id]
            traverse(t.children_left[node_id], low, th)
            traverse(t.children_right[node_id], th, high)

    traverse(0, a, b)

    # Интеграл = сумма (среднее в листе) * (длина интервала)
    integral = 0.0
    for low, high, val in leaves:
        integral += val * (high - low)
    return integral

# ---------- Сравнение точности в зависимости от числа отрезков M ----------
M_values = np.logspace(1, 2.6, 15, dtype=int)  # от 10 до ~400

# Для хранения ошибок
errors_tree_f1 = []
errors_left_f1 = []
errors_right_f1 = []
errors_mid_f1 = []

errors_tree_f2 = []
errors_left_f2 = []
errors_right_f2 = []
errors_mid_f2 = []

for M in M_values:
    # ---------- Функция 1 ----------
    # Дерево
    I_tree = tree_integral(f1, a1, b1, M, N_train=5000)
    errors_tree_f1.append(abs(I_tree - I1_exact))

    # Прямоугольники
    I_left = rectangle(f1, a1, b1, M, 'left')
    I_right = rectangle(f1, a1, b1, M, 'right')
    I_mid = rectangle(f1, a1, b1, M, 'mid')
    errors_left_f1.append(abs(I_left - I1_exact))
    errors_right_f1.append(abs(I_right - I1_exact))
    errors_mid_f1.append(abs(I_mid - I1_exact))

    # ---------- Функция 2 ----------
    I_tree = tree_integral(f2, a2, b2, M, N_train=5000)
    errors_tree_f2.append(abs(I_tree - I2_exact))

    I_left = rectangle(f2, a2, b2, M, 'left')
    I_right = rectangle(f2, a2, b2, M, 'right')
    I_mid = rectangle(f2, a2, b2, M, 'mid')
    errors_left_f2.append(abs(I_left - I2_exact))
    errors_right_f2.append(abs(I_right - I2_exact))
    errors_mid_f2.append(abs(I_mid - I2_exact))

# ---------- Визуализация ----------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# График для f1
ax1.loglog(M_values, errors_tree_f1, 'o-', label='Дерево решений')
ax1.loglog(M_values, errors_left_f1, 's--', label='Левые прямоуг.')
ax1.loglog(M_values, errors_right_f1, '^--', label='Правые прямоуг.')
ax1.loglog(M_values, errors_mid_f1, 'd--', label='Центральные прямоуг.')
ax1.set_xlabel('Число отрезков M')
ax1.set_ylabel('Абсолютная погрешность')
ax1.set_title(r'$f(x)=4/(1+x^2)$ на $[0,1]$')
ax1.legend()
ax1.grid(True, which='both', ls=':')

# График для f2
ax2.loglog(M_values, errors_tree_f2, 'o-', label='Дерево решений')
ax2.loglog(M_values, errors_left_f2, 's--', label='Левые прямоуг.')
ax2.loglog(M_values, errors_right_f2, '^--', label='Правые прямоуг.')
ax2.loglog(M_values, errors_mid_f2, 'd--', label='Центральные прямоуг.')
ax2.set_xlabel('Число отрезков M')
ax2.set_ylabel('Абсолютная погрешность')
ax2.set_title(r'$f(x)=(\sin x)^x(\ln\sin x + x\,\mathrm{ctg}\,x)$ на $[0,\pi]$')
ax2.legend()
ax2.grid(True, which='both', ls=':')

plt.tight_layout()
plt.show()
```

```python
import json
import numpy as np

# Загрузка данных из JSON-файла (предположим, файл называется data.json)
with open("data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Извлечение признаков и целевых переменных
X = np.array([item["both_embedding"] for item in data])  # (N, D)
y = np.array([[item["citations"], item["downloads"]] for item in data])  # (N, 2)

# Опционально: можно нормализовать X (привести строки к единичной длине)
# Это не меняет косинусное расстояние, но упрощает вычисления
X_norm = X / np.linalg.norm(X, axis=1, keepdims=True)

def cosine_distance(x1, x2, assume_normalized=False):
    """
    Косинусное расстояние: d(x1,x2) = 1 - (x1·x2)/(||x1||*||x2||).
    Если assume_normalized=True, то векторы уже нормализованы и формула упрощается.
    """
    if assume_normalized:
        # Для нормализованных векторов расстояние = 1 - скалярное произведение
        cos_sim = np.dot(x1, x2)
        return 1.0 - cos_sim
    else:
        # Общий случай
        norm1 = np.linalg.norm(x1)
        norm2 = np.linalg.norm(x2)
        if norm1 == 0 or norm2 == 0:
            return 1.0  # Если один из векторов нулевой, максимальное расстояние
        cos_sim = np.dot(x1, x2) / (norm1 * norm2)
        return 1.0 - cos_sim

def knn_predict(query, X_train, y_train, k, assume_normalized=True, eps=1e-10):
    """
    Предсказание методом kNN с косинусным расстоянием и взвешенным средним.

    Параметры:
    query: одномерный массив признаков запроса (D,)
    X_train: обучающая выборка признаков (N, D)
    y_train: обучающие целевые значения (N, 2) – цитирования/скачивания
    k: число ближайших соседей
    assume_normalized: если True, то векторы считаются предварительно нормализованными
    eps: маленькая константа для избежания деления на ноль в весах (если расстояние=0)

    Возвращает:
    предсказанный вектор (2,) – (citations, downloads)
    """
    # Вычисляем косинусные расстояния от запроса до всех точек обучения
    if assume_normalized:
        # Для нормализованных векторов расстояние = 1 - скалярное произведение
        similarities = np.dot(X_train, query)   # (N,)
        distances = 1.0 - similarities
    else:
        distances = np.array([cosine_distance(query, x, assume_normalized=False) for x in X_train])

    # Находим k ближайших соседей (индексы)
    k_indices = np.argpartition(distances, k)[:k]
    k_distances = distances[k_indices]

    # Вычисляем веса: ω_i = 1 / (d_i + eps)
    weights = 1.0 / (k_distances + eps)

    # Взвешенное среднее целевых значений
    y_neighbors = y_train[k_indices]  # (k, 2)
    weighted_y = np.sum(weights[:, np.newaxis] * y_neighbors, axis=0)  # (2,)
    sum_weights = np.sum(weights)

    return weighted_y / sum_weights

# ================== Демонстрация ==================
# Используем первую статью как тестовый запрос, исключим её из обучения
query_idx = 9
query_vector = X_norm[query_idx]          # нормализованный
y_true = y[query_idx]

# Обучающая выборка без этой статьи
train_indices = list(range(len(X)))
train_indices.remove(query_idx)
X_train_norm = X_norm[train_indices]
y_train = y[train_indices]

# Параметры
k = 5
eps = 1e-10

prediction = knn_predict(query_vector, X_train_norm, y_train, k, assume_normalized=True, eps=eps)

print(f"Истинные значения: citations={y_true[0]}, downloads={y_true[1]}")
print(f"Предсказанные значения (k={k}): citations={prediction[0]:.2f}, downloads={prediction[1]:.2f}")
print(f"Относительная ошибка по citations: {abs(prediction[0] - y_true[0]) / (y_true[0] + 1e-9):.4f}")
print(f"Относительная ошибка по downloads: {abs(prediction[1] - y_true[1]) / (y_true[1] + 1e-9):.4f}")
```
