```python
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1. Генерация данных (100 точек, шум sigma=0.5)
# ------------------------------------------------------------
np.random.seed(42)
N = 100
x = np.linspace(0, 10, N)

A1_true, om1_true, ph1_true = 1.0, 2.0, 0.3
A2_true, om2_true, ph2_true = 0.7, 5.0, 1.0
sigma = 0.5

y_true = (A1_true * np.sin(om1_true * x + ph1_true) +
          A2_true * np.sin(om2_true * x + ph2_true))
y_obs = y_true + np.random.normal(0, sigma, size=N)

# ------------------------------------------------------------
# 2. Байесовская модель
# ------------------------------------------------------------
with pm.Model() as model:
    # Априорные распределения
    A1 = pm.Uniform('A1', lower=0, upper=3)
    om1 = pm.Uniform('om1', lower=0, upper=10)
    ph1 = pm.Uniform('ph1', lower=-np.pi, upper=np.pi)

    A2 = pm.Uniform('A2', lower=0, upper=3)
    om2 = pm.Uniform('om2', lower=0, upper=10)
    ph2 = pm.Uniform('ph2', lower=-np.pi, upper=np.pi)

    # Модель
    mu = (A1 * pm.math.sin(om1 * x + ph1) +
          A2 * pm.math.sin(om2 * x + ph2))

    # Правдоподобие (известная sigma)
    y = pm.Normal('y', mu=mu, sigma=sigma, observed=y_obs)

# ------------------------------------------------------------
# 3. Сэмплирование SMC (гомотопия, множество уолкеров)
# ------------------------------------------------------------
with model:
    trace = pm.sample_smc(
        draws=2000,
        chains=4,
        random_seed=42,
        progressbar=True
    )

# ------------------------------------------------------------
# 4. Вычисляем логарифмическое правдоподобие для каждого сэмпла
# ------------------------------------------------------------
with model:
    trace = pm.compute_log_likelihood(trace)

# ------------------------------------------------------------
# 5. Анализ результатов
# ------------------------------------------------------------
# Апостериорные маргиналы
az.plot_trace(trace, var_names=['A1', 'om1', 'ph1', 'A2', 'om2', 'ph2'])
plt.tight_layout()
plt.savefig('trace_smc.png', dpi=150)
plt.show()

# Сводка
summary = az.summary(trace, var_names=['A1', 'om1', 'ph1', 'A2', 'om2', 'ph2'])
print("Апостериорные оценки (медиана и 94% HDI):")
print(summary)

# Наиболее вероятные значения (MAP / MLE) – максимум логарифмической правдоподобия
log_lik = trace.log_likelihood['y'].sum(dim='y_dim_0')   # сумма log P(y_i|theta) по всем точкам
idx_map = log_lik.argmax()                               # индекс с максимальным значением
map_sample = {}
for var in ['A1', 'om1', 'ph1', 'A2', 'om2', 'ph2']:
    # Превращаем posterior (chain, draw) в одномерный массив
    flat_samples = trace.posterior[var].values.flatten()
    map_sample[var] = flat_samples[idx_map.values]

print("\nНаиболее вероятные значения параметров (MAP/MLE):")
for k, v in map_sample.items():
    print(f"  {k} = {v:.4f}")

# ------------------------------------------------------------
# 6. Проверка по критерию хи-квадрат
# ------------------------------------------------------------
y_pred = (map_sample['A1'] * np.sin(map_sample['om1'] * x + map_sample['ph1']) +
          map_sample['A2'] * np.sin(map_sample['om2'] * x + map_sample['ph2']))

chi2 = np.sum(((y_obs - y_pred) / sigma) ** 2)
dof = N - 6
chi2_red = chi2 / dof

print(f"\nχ² = {chi2:.2f} при {dof} степенях свободы (ожидаемое ~ {dof} ± {np.sqrt(2*dof):.1f})")
print(f"Приведённый χ² = {chi2_red:.3f} (в идеале ~ 1.0)")

# График
plt.figure(figsize=(10, 4))
plt.plot(x, y_obs, 'k.', alpha=0.6, label='Данные (с шумом)')
plt.plot(x, y_true, 'b-', lw=2, label='Истинная кривая')
plt.plot(x, y_pred, 'r--', lw=2, label=f'MAP (χ²={chi2:.1f})')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Сумма двух синусоид: данные и восстановление')
plt.savefig('fit_smc.png', dpi=150)
plt.show()
```
